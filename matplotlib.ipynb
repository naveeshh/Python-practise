# Title

__title(label,fontdict=None,loc=None,pad=None,**kwrgs)__

```python
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("First plot",{"size":20,"color":"blue"},loc="left",pad=20)
```

# Grid

__grid(b=None,which='major',axis='both',**kwrgs)__

```python
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("First plot")
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("First plot")
plt.minorticks_on()
plt.grid(b=True,which='minor',color='red')
plt.grid(b=True,which='major',color='green')

plt.show()
```

```python
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("First plot")
plt.grid(b=True,axis='x')
plt.show()
```

```python
plt.plot([1,2,3,4],[5,6,7,8])
left,right=plt.xlim()
print(left,right)
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("First plot")
plt.show()
```

```python
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlim(2,6)
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("First plot")

plt.show()
```

# Legend

```python
t=np.array([1,2,3,4])
plt.plot(t**2,t,color='red',label='square')
plt.plot(t**3,t,color='green',label='cube')

plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("First plot")
plt.legend()
plt.show()
```

```python
t=np.array([1,2,3,4])
plt.plot(t**2,t,color='red')
plt.plot(t**3,t,color='green')

plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("First plot")
plt.legend(['square'])
plt.show()
```
