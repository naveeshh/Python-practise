# Types of Arguments

1. Formal Arguments
2. Actual Arguments------>  1.position
                            2.Keyword
                            3.Default
                            4.variable length

### position

We have to follow the order of the position which we given in the formal an actual parameters

```python
def person(name,age):
    print(name)
    print(age)
person("naveesh",21)    
```

### keyword

Here we give keyword to actual parameters

```python
def person(name,age):
    print(name)
    print(age)
person(age=28,name="naveesh") 
```

### default 

Here we define values to formal parameters only.
If we forgot  to write value in actual parameter it takes default value

```python
def person(name,age=10):
    print(name)
    print(age)
person("naveesh",21) 
```

### Variale length

*y represents y takes multiple values

```python
def sum(x,*y):
    c=x+y
    print(c)
sum(4,5,7,8,2)    
```

```python
def sum(x,*y):
    print(x)
    print(y)
    c=x
    for i in y:
        c=c+i
    print(c)
sum(4,5,7,8,2)
```

# Scope

In order to make a local variable as local variable   make "global x" in the function

```python
a=10
def nav():
    a=15
    print("local a",a)
nav()
print("outsie a",a)
```

```python
a=10
def nav():
    global a
    a=15
    print("local a",a)
nav()
print("outsie a",a)
```

# Lambda

Functions without name is called anoynous function. It is also called lambda 

```python
def square(lst):
    for i in lst:
        print(i*i)
lst=(2,6,8)
square(lst)
```

```python
f=lambda a: a*a
result=f(5)
print(result)
```

# Filter ,Reduce, Map 

### filter takes two arguments
     filter(func,seq)
     reduce(func,seq)

```python
from functools import reduce
nums=[3,87,89,90,45,4,1]
evens=list(filter(lambda x: x%2 == 0,nums))
double=list(map(lambda x: x*2,evens))
sum=reduce(lambda a,b:a+b,double)
print(sum)
print(double)
print(evens)
```

# Decoraters

When we want to make a change in the function without touching the function.
using Decoraters we can ad exra features to the function without touching the function

```python
def div(a,b):
    print(a/b)
def smart_div(func):
    def inner(a,b):
        if a<b:
            a,b=b,a
        return func(a,b)
    return inner
div=smart_div(div)
div(2,4)    
```
